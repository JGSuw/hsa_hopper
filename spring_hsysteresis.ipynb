```python
import numpy as np
from matplotlib import pyplot as plt

t = np.arange(start=0.,stop=1.,step=1e-3)
omega = 2*np.pi
x = 1-np.cos(omega*t)
xdot = omega*np.sin(omega*t)
y = np.zeros(len(x))
for i in range(len(x)):
    if xdot[i]>= 0:
        y[i] = 10*x[i]
    else:
        y[i] = 10*x[i]-5/(omega)**2*xdot[i]**2

plt.plot(1-x,y)

dy = np.diff(y)
dx = np.diff(x)
np.average(dy/dx)
```
