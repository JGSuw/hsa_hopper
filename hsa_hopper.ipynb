```python
import numpy as np
from matplotlib import pyplot as plt
```

Let's consider the properties we want in a hopper by first examing a model of hopping dynamics based on a constant-coefficient linear equation,

$$ \ddot{x} + \omega^2 x + g - \omega^2 l = 0 $$

we assume that the maximum compression of the spring element is $\delta x$, and we denote the height gain that can be achieved by a hop as $\delta h$. An upper bound on the height gain is readily derived from the (normalized) energy expression,

$$ g \delta h \leq \cfrac{\omega^2}{2}\delta x^2 - g\delta x $$

where the datum for gravitational potential is fixed at $x = l$. As well, $\omega$ has an absolute lower bound,

$$ \sqrt{\cfrac{2g}{\delta x}} < \omega$$

which is required so that $\delta h > 0$

```python
# now to demonstrate this basic relationship between hopping height, natural frequency, and spring displacement
delta_x = [0.03, 0.04, 0.05, 0.06, 0.07]
omega = [np.linspace(np.sqrt(2*9.81/x),10*np.pi,num=100) for x in delta_x]
delta_h = [np.array([(w**2)/2/9.81*delta_x[i]**2-delta_x[i] for w in omega[i]]) for i in range(len(delta_x))]

fig = plt.plot()
for i in range(len(delta_x)):
    plt.plot(omega[i]/2/np.pi, delta_h[i]/.01,label="delta_x = %f" % delta_x[i])
plt.title("Theoretical maximum hopping height")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Height gain (cm)")
plt.legend(["delta x = %.1f cm" % (100*x) for x in delta_x])
plt.savefig("max height vs frequency")

```

```python
# same plot, but with mass on the x axis, assuming K = 500
delta_x = [0.03, 0.04, 0.05, 0.06, 0.07]
# omega = 500
omega = [np.linspace(np.sqrt(2*9.81/x),10*np.pi,num=100) for x in delta_x]
delta_h = [np.array([(w**2)/2/9.81*delta_x[i]**2-delta_x[i] for w in omega[i]]) for i in range(len(delta_x))]

fig = plt.plot()
for i in range(len(delta_x)):
    plt.plot(700/omega[i]**2, delta_h[i]/.01,label="delta_x = %f" % delta_x[i])
plt.title("Theoretical maximum hopping height (K=700 N/m)")
plt.xlabel("Payload mass (Kg)")
plt.ylabel("Height gain (cm)")
plt.legend(["x = %.1f cm" % (100*x) for x in delta_x])
# plt.hlines([(700/2/9.81*.06**2-.06)/.01],xmin=[0.7],xmax=[1.0])
# plt.vlines([1.0],ymin=[0.],ymax=[(700/2/9.81*.06**2-.06)/.01])
plt.scatter([1.0],[(700/2/9.81*.06**2-.06)/.01],marker="D",color="red")
fig
plt.savefig("max height vs payload mass")
```
